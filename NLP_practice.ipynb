```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import tensorflow_datasets as tfds

data, info = tfds.load("imdb_reviews", with_info=True, as_supervised=True)

```

```python
train, test = data["train"], data["test"]

train_s = []
train_l = []

test_s = []
test_l = []

for s,l in train:
  train_s.append(s.numpy().decode("utf-8"))
  train_l.append(l.numpy())

for s,l in test:
  test_s.append(s.numpy().decode("utf-8"))
  test_l.append(l.numpy())

train_l = np.array(train_l)
test_l = np.array(test_l)
```

```python
vocab_size = 10000
oov_token = "<OOV>"
trunc_type = "post"
max_len = 128
embedding_dim = 16

tokenizer = Tokenizer(vocab_size, oov_token)
tokenizer.fit_on_texts(train_s)
train_seq = tokenizer.texts_to_sequences(train_s)
pad_train_seq = pad_sequences(train_seq, truncating=trunc_type, maxlen=max_len)
test_seq = tokenizer.texts_to_sequences(test_s)
pad_test_seq = pad_sequences(test_seq, truncating=trunc_type, maxlen=max_len)



```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_len),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(32, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
model.fit(pad_train_seq,train_l, batch_size=32, epochs=10, validation_data=(pad_test_seq,test_l))
```
